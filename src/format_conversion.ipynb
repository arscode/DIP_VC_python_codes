## RGB to Grayscale 

```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

testImage = cv2.imread('./img/yoda.jpg')
grayImage = cv2.cvtColor(testImage, cv2.COLOR_BGR2GRAY)

#cv2.imshow('Original image', testImage)
#cv2.imshow('Gray image', grayImage)

# this part is to make the pictures visible in jupyter notebook
plt.subplot(2,2,1)
plt.imshow(testImage)
plt.subplot(2,2,2)
plt.imshow(grayImage, cmap='gray')
plt.show()
```

## Gray to binary

```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

test_gray_image = cv2.imread('./img/gray_dog.jpg', cv2.IMREAD_GRAYSCALE)
thresh = 128
img_binary = cv2.threshold(test_gray_image, thresh, 255, cv2.THRESH_BINARY)[1]
cv2.imwrite('./img/binary_dog.jpg',img_binary) 


# mismatch due to imshow here called from matplotlib, not form opencv
plt.subplot(2,2,1)
plt.imshow(test_gray_image, cmap='gray')
plt.subplot(2,2,2)
plt.imshow(img_binary, cmap='gray')
plt.show()
```

## RGB to binary


```python
import cv2

# take the rgb image and convert into grayscale
rgb_image = cv2.imread('./img/yoda.jpg', 2)
(thresh, img_binary) = cv2.threshold(rgb_image, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

cv2.imwrite('./img/yoda_bin.jpg', img_binary)

# this part is to make the pictures visible in jupyter notebook
plt.subplot(2,2,1)
plt.imshow(rgb_image, cmap='gray')
plt.subplot(2,2,2)
plt.imshow(img_binary, cmap='gray')
plt.show()
```

## RGB to HSV

```python
import cv2

rgb_image = cv2.imread('./img/yoda.jpg')
hsv_img = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2HSV)

cv2.imwrite('./img/yoda_hsv.jpg', hsv_img)
```

## HSV to RGB

```python
import cv2

hsv_img = cv2.imread('./img/yoda_hsv.jpg')
rgb_img = cv2.cvtColor(hsv_img, cv2.COLOR_HSV2BGR)

cv2.imwrite('./img/yoda_rgb.jpg', rgb_image)
```

## RGB to YCbCr

```python
import cv2

rgb_image = cv2.imread('./img/yoda.jpg')
YCbCr_img = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2YCrCb)

# saving file
cv2.imwrite('./img/yoda_YCbCr.jpg', YCbCr_img)
```

## YCbCr to RGB

```python
import cv2

YCbCr_image = cv2.imread('./img/yoda_YcbCr.jpg')
rgb_img = cv2.cvtColor(YCbCr_image, cv2.COLOR_YCrCb2BGR)

# saving file
cv2.imwrite('./img/yoda_YCbCr2RGB.jpg', rgb_img)

```

## Rotating image clockwise and anti-clockwise

### clockwise rotation

```python
import cv2

# rotate by 90 degree clockwise
src_img = cv2.imread('./img/yoda.jpg')
rot_img = cv2.rotate(src_img, cv2.ROTATE_90_CLOCKWISE)

cv2.imwrite('./img/yoda_90degRotClockwise.jpg', rot_img)
```

```python
import cv2

src_img = cv2.imread('./img/yoda.jpg')
rot_img = cv2.rotate(src_img, cv2.ROTATE_180)

cv2.imwrite('./img/yoda_180degRot.jpg', rot_img)
```

### anti-clockwise rotation

```python
import cv2

src_img = cv2.imread('./img/yoda.jpg')
rot_img = cv2.rotate(src_img, cv2.ROTATE_90_COUNTERCLOCKWISE)

cv2.imwrite('./img/yoda_90degRotANTICLOCKWISE.jpg', rot_img)
```

# Find histogram of a gray image using openCV

```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

# loading image in grayscale
src_img = cv2.imread('./img/yoda.jpg', 0)

# calculate frequency of pixels in range 0-255 
histg = cv2.calcHist([src_img],[0],None,[256],[0,256])

plt.subplot(2,2,1)
plt.imshow(src_img, cmap='gray')
plt.subplot(2,2,2)
plt.plot(histg)

# alternative way
#plt.hist(src_img.ravel(),256,[0,256]) 
plt.show()
```
